# Graphik und Animatonen mit der Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Einfacher Graph mit Listen

```python
lx, ly = [], []
for x in range(11):
    y = x**2
    lx.append(x)
    ly.append(y)
    
plt.plot(lx, ly)
```

## Graph mit NumPy-Arrays ohne Schleifen

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 100)
y = x**2
plt.plot(x, y)
```

## Gitternetzlinien

```python
t = np.linspace(0, 20, 500)
u = 325*np.sin(2*np.pi*50*t/1000)
```

```python
plt.plot(t, u, linewidth = 2)
plt.grid()
```
